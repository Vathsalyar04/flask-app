```python
# Import required libraries
import numpy as np
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense

# Define dataset
X = np.array([[4], [5], [6], [7], [8], [9], [10]], dtype=np.float32)
y = np.array([8, 10, 12, 14, 16, 18, 20], dtype=np.float32)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=7)

print("Features Training Data:", X_train)
print("Labels Training Data:", y_train)
print("Features Test Data:", X_test)
print("Labels Test Data:", y_test)

# Define the model
model = Sequential([
    Dense(64, input_shape=(1,), activation="relu"),
    Dense(32, activation="relu"),
    Dense(16, activation="relu"),
    Dense(1, activation="linear")  # Linear activation for regression
])

# Compile the model
model.compile(loss="mean_squared_error", optimizer="adam", metrics=["mean_squared_error"])

# Train the model
model.fit(X_train, y_train, epochs=200, verbose=1)

# Make predictions
X_test = np.array(X_test, dtype=np.float32)  # Ensure X_test is a NumPy array
predictions = model.predict(X_test)

print("Predictions:")
for i in range(len(X_test)):
    print(f"Input: {X_test[i][0]}, Predicted Output: {predictions[i][0]}")

# Evaluate the model
score = model.evaluate(X_test, y_test, verbose=1)

# Print evaluation metrics
for i in range(len(model.metrics_names)):
    print(f"Metric {model.metrics_names[i]}: {round(score[i], 2)}")
```
